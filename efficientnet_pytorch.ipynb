```python
def load_efficientnet(path, model_name='efficientnet-b7', pretrained=True, num_classes=1000):
    assert model_name in [f'efficientnet-b{i}' for i in range(8)]
    model = EfficientNet.from_name(model_name, override_params={'num_classes':num_classes})

    if pretrained:
        model_state = torch.load(path)
        # load original weights apart from its head
        if '_fc.weight' in model_state.keys():
            model_state.pop('_fc.weight')
            model_state.pop('_fc.bias')
            res = model.load_state_dict(model_state, strict=False)
            assert str(res.missing_keys) == str(['_fc.weight', '_fc.bias']), 'issue loading pretrained weights'
        else:
            # A basic remapping is required
            from collections import OrderedDict
            mapping = { i:o for i,o in zip(model_state.keys(), model.state_dict().keys()) }
            mapped_model_state = OrderedDict([
                (mapping[k], v) for k,v in model_state.items() if not mapping[k].startswith('_fc')
            ])
            res = model.load_state_dict(mapped_model_state, strict=False)
            print(res)
    return model
```

```python
class EfficientNetClassification(nn.Module):
    def __init__(
            self,
            model_name='effiecientnet-b5',
            model_path=None,
            pretrained=True,
            fcs=(512, 256),
            num_classes=1000,
            p_dropout=0.5,
    ):
        super(EfficientNetClassification, self).__init__()

        assert model_path is not None or pretrained is False, 'pretrained model require model_path'
        self.efficient = load_efficientnet(
            path=model_path,
            model_name=model_name,
            pretrained=pretrained,
            num_classes=20,
        )
        
        self._avg_pooling = nn.AdaptiveAvgPool2d(1)

        self._mlp = self._get_mlp(
            input_size=self.efficient._fc.in_features,
            hidden_sizes=fcs,
            num_classes=num_classes,
            p_dropout=p_dropout,
        )

    def forward(self, inputs):
        bs = inputs.size(0)
        x = self.efficient.extract_features(inputs)
        x = self._avg_pooling(x)
        x = x.view(bs, -1)
        return self._mlp(x)


    def extract_features(self, inputs):
        return self.efficient.extract_features(inputs)

    @staticmethod
    def _get_mlp(input_size, hidden_sizes, num_classes, p_dropout):
        layer_sizes = [input_size, *hidden_sizes, num_classes]

        layers = []
        for i in range(len(layer_sizes) - 1):
            if i != len(layer_sizes) - 1:
                layers.extend([
                    nn.Dropout(p=p_dropout),
                    nn.Linear(layer_sizes[i], layer_sizes[i + 1]),
                    # nn.Sigmoid(),
                    MemoryEfficientSwish(),
                ])
            else:
                layers.extend([
                    nn.Dropout(p=p_dropout),
                    nn.Linear(layer_sizes[i], layer_sizes[i + 1]),
                ])

        return nn.Sequential(*layers)
```

```python
class FocalLoss(nn.Module):
    def __init__(self, gamma=3., reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = nn.CrossEntropyLoss(reduction='none')(inputs, targets)
        pt = torch.exp(-ce_loss)
        f_loss = ((1 - pt)**self.gamma) * ce_loss
        if self.reduction == 'sum':
            return f_loss.sum()
        elif self.reduction == 'mean':
            return f_loss.mean()
```

```python
top_3_accuracy = partial(top_k_accuracy, k=3)
top_3_accuracy.__name__ = 'top_3_accuracy'
top_5_accuracy = partial(top_k_accuracy, k=5)
top_5_accuracy.__name__ = 'top_5_accuracy'
learn = Learner(
    data=data,
    model=model,
    path=train_model_path,
    loss_func=FocalLoss(),
    metrics=[
        accuracy,
        top_3_accuracy,
        top_5_accuracy,
    ],
    callback_fns=[
        BnFreeze,
        # OverSamplingCallback,
        # partial(GradientClipping, clip=0.2),
        # partial(SaveModelCallback, monitor='kappa_score', name='best_kappa')
    ]
)
# learn.model_dir = train_model_path
learn.split(lambda m: (model._conv_head,) )
learn.freeze()
```

```python
learn.freeze()
learn.lr_find(start_lr=1e-5, end_lr=1e1, wd=5e-3)
learn.recorder.plot(suggestion=True)
```
